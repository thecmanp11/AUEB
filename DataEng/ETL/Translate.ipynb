```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
import re
import csv

options = Options()
options.headless = True
driver = webdriver.Chrome(options=options, executable_path=r'/usr/local/bin/chromedriver')
driver.implicitly_wait(5)

csv_fs = ['betshop']

data = []

# Load scraped matches
for csv_f in csv_fs:

    with open(csv_f+'.csv') as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')

        for row in csv_reader:

            data.append(row)

# Load Cached Translations
translations = {}

with open("translation.csv") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        translations[row[0]] = row[1]

# Visit Website
driver.get("https://translate.google.gr/")
time.sleep(1)

# Click I Agree
driver.find_elements(By.TAG_NAME, "button")[0].click()
time.sleep(2)

# Reload Page
driver.get("https://translate.google.gr/")
time.sleep(1)

# Perform the translation
for r in data:
    
    home = r[2]
    
    # Load home
    driver.get("https://translate.google.gr/?sl=el&tl=en&text={}&op=translate".format(home))
    time.sleep(2)

    divs = driver.find_elements(By.TAG_NAME, "div")
    
    res = ""
    for div in divs:
        
        if "Αποτελέσματα μετάφρασης" in div.text:
            res = div.text
            break
    
    translated = res.split("Αποτελέσματα μετάφρασης")[1].split("volume_up")[0].strip()

    print("{} => {}".format(home,translated))

driver.quit()
```
