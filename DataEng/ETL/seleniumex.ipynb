```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import time
import json
import re
import csv
import re
import os, errno

options = Options()
# options.headless = True
driver = webdriver.Chrome(options=options, executable_path=r'/usr/local/bin/chromedriver')
driver.implicitly_wait(5)

# Save output on a file
def savePage(path,filename,data):

    # Build the folder path if it does not exist
    try:
        os.makedirs(path)
    except OSError as e:
        if e.errno != errno.EEXIST:
            raise

    # Write data file
    full_file_path = path + filename
    json_file = open(full_file_path,'w')
    json_file.write(data)
    json_file.close()
    return json_file



for i in range(0,198):
    
    driver.get("https://www.11888.gr/yellow-pages/logistika-grafeia/?query=%CE%9B%CE%BF%CE%B3%CE%B9%CF%83%CF%84%CE%B9%CE%BA%CE%AC-%CE%93%CF%81%CE%B1%CF%86%CE%B5%CE%AF%CE%B1&page={}".format(i))
    
    results = driver.execute_script("return (JSON.stringify(window.vueData.results));")

    savePage("./results/","page_{}.json".format(i),results)

    print("Saved Page: {}".format(i+1))

driver.quit()
```

```python
import json, sys

# Generate a file list ..

# Load file list
file_list = open('./file.list', 'r') 
lines = file_list.readlines() 
dataset = []

for line in lines:
    with open(line.strip(), 'r') as data_file:
        offices = data_file.read().replace('\n', '')
        
        offices = json.loads(offices)
        
        for office in offices:

            data = {
                'keywords': [],
                'displayName': office.get('displayName'),
                'companyName': office.get('companyName'),
                'description_short': "",
                'description_long': "",
                'phones': [],
                'isActive': office.get('isActive'),
                'contactTime_start': office.get('contactTime').get('start'),
                'contactTime_end': office.get('contactTime').get('end'),
                'website': office.get('website'),
                'region': office.get('address').get('region').get('name'),
                'subregion': office.get('address').get('subregion').get('name'),
                'municipality': office.get('address').get('municipality', {}).get('name'),
                'full_address': office.get('address').get('description'),
                'street1': office.get('address').get('street1'),
                'number1': office.get('address').get('number1'),
                'lon': office.get('address').get('location', {}).get('lon'),
                'lat': office.get('address').get('location', {}).get('lat'),
                'postal': office.get('address').get('postal_code'),
                'detailUrl': office.get('detailUrl'),
                'status': office.get('status'),
                'serviceAreasAddons': office.get('serviceAreasAddons')
            }

            if office.get('keywords') is not None:
                for keyword in office['keywords']:
                    data['keywords'].append(keyword['name'])
            data['keywords'] = ",".join(data['keywords'])

            if office.get('phones') is not None:
                for phone in office['phones']:
                    data['phones'].append(phone['number'])
            data['phones'] = ",".join(data['phones'])

            data['description_short'] = office.get('description').get('short')
            if data['description_short'] is not None:
                data['description_short'] = data['description_short'].replace('"','').replace("'",'').replace('\n','')

            data['description_long'] = office.get('description').get('long')
            if data['description_long'] is not None:
                data['description_long'] = data['description_long'].replace('"','').replace("'",'').replace('\n','')

            dataset.append(data)
    
    data_file.close()

file_list.close()

office
```
