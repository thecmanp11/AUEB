```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
import re
import csv

site_name = "betshop"

options = Options()
# options.headless = True
# MAC chromedriver error: xattr -d com.apple.quarantine chromedriver
driver = webdriver.Chrome(options=options, executable_path=r'/usr/local/bin/chromedriver')
driver.implicitly_wait(60)

# Visit Website
driver.get("https://www.betshop.gr/sports")
time.sleep(1)

# Click on cookies
driver.find_element_by_id("onetrust-accept-btn-handler").click()
time.sleep(1)

# View full coupon
driver.get("https://www.betshop.gr/sports/game/stoixima-podosfairo/1/coupon")
time.sleep(1)

days = driver.find_elements_by_css_selector(".flex.ml-4 > div")
for day in days:
    
    day.click()

    time.sleep(2)

    sportEvents = driver.find_element_by_id("sportEvents")
    
    matches = sportEvents.find_elements_by_css_selector(".group")
    
    for match in matches:
        
        time_league_name_block = match.find_element_by_css_selector(".w-4\\/12")
        
        time_league = time_league_name_block.find_element_by_css_selector(".no-wrap").find_elements_by_css_selector(".text-sm")
        
        print("Time: {}".format(time_league[0].text))
        print("League: {}".format(time_league[1].text))

        name_blocks = time_league_name_block.find_elements_by_css_selector(".truncate")
        
        print("Home: {}".format(name_blocks[1].text))
        print("Away: {}".format(name_blocks[2].text))

        all_odds = match.find_elements_by_css_selector(".w-8\\/12")
        
        odds_1X2 = all_odds[0].find_element_by_css_selector(".flex.w-full.text-center.mr-1").find_elements_by_css_selector(".mr-1")
        
        print("Odds for 1: {}".format(odds_1X2[0].text))
        print("Odds for X: {}".format(odds_1X2[1].text))
        print("Odds for 2: {}".format(odds_1X2[2].text))
        
        line = ["betshop", time_league[0].text, time_league[1].text, name_blocks[0].text, name_blocks[1].text,
               odds_1X2[0].text,
               odds_1X2[1].text,
               odds_1X2[2].text]
        
        print(line)

        with open("results/betshop.csv", 'a', newline='') as myfile:
            wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
            wr.writerow(line)

driver.quit()
```
